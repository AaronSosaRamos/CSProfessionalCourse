#Review of Probabilities: Random Variable, Probability Function, Expected Value

* Random Variable: In probability theory, a random variable is a variable whose possible values are outcomes of a random phenomenon. In the provided code, the random_variable function simulates the random variable by taking a dictionary of values and their corresponding probabilities. It returns one of the values based on their probabilities.

* Probability Function: The probability function assigns probabilities to each outcome of a random variable. In the example, the probabilities dictionary holds the values and their associated probabilities.

* Expected Value: The expected value of a random variable represents the average value it would take over an infinite number of trials. It is calculated as the sum of each value multiplied by its probability. In the provided code, this is not explicitly calculated, but it's a fundamental concept in probability theory.

```python
# Importing required libraries
import random

# Defining a function to simulate a random variable
def random_variable(probabilities):
    rand_num = random.random()
    cumulative_prob = 0
    for value, prob in probabilities.items():
        cumulative_prob += prob
        if rand_num <= cumulative_prob:
            return value

# Example usage
probabilities = {'A': 0.3, 'B': 0.5, 'C': 0.2}
sample = random_variable(probabilities)
print("Sample drawn from random variable:", sample)
```

#Hiring Problem

* Hiring Problem: In the hiring problem, we are given a sequence of candidates one by one, and we need to select the best one. The objective is to maximize the probability of selecting the best candidate.

* Implementation: The hiring_problem function simulates the hiring problem for n applicants. Each applicant is assigned a random score (representing their quality), and the function selects the candidate with the highest score.

```python
# Defining a function to simulate the hiring problem
def hiring_problem(n):
    best_candidate = -1
    best_score = float('-inf')
    for i in range(n):
        candidate_score = random.random()  # Simulating candidate's quality
        if candidate_score > best_score:
            best_candidate = i
            best_score = candidate_score
    return best_candidate

# Example usage
num_applicants = 10
best_candidate_index = hiring_problem(num_applicants)
print("Best candidate index:", best_candidate_index)
```

```python
import random
import matplotlib.pyplot as plt

def hiring_problem(n):
    best_candidate = -1
    best_score = float('-inf')
    for i in range(n):
        candidate_score = random.random()  # Simulating candidate's quality
        if candidate_score > best_score:
            best_candidate = i
            best_score = candidate_score
    return best_candidate

# Number of simulations for each data length
num_simulations = 1000

# Data lengths to analyze
data_lengths = [10, 50, 100, 500]

# Perform simulations for each data length
results = []
for length in data_lengths:
    best_candidates = []
    for _ in range(num_simulations):
        best_candidates.append(hiring_problem(length))
    results.append(best_candidates)

# Plotting the results
plt.figure(figsize=(10, 6))
for i in range(len(data_lengths)):
    plt.hist(results[i], bins=range(data_lengths[i]+1), alpha=0.5, label=f'n = {data_lengths[i]}')

plt.xlabel('Index of Best Candidate')
plt.ylabel('Frequency')
plt.title('Distribution of Best Candidate Index')
plt.legend()
plt.grid(True)
plt.show()
```

#Quicksort Algorithm. Quicksort Time Analysis: Worst Case, Best Case, and Average Case

```python
import random
import time

def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

def generate_random_array(length):
    return [random.randint(1, 1000) for _ in range(length)]

def time_analysis(data_lengths):
    time_worst_case = []
    time_best_case = []
    time_average_case = []

    for length in data_lengths:
        # Generating arrays for worst, best, and average cases
        arr_worst_case = list(range(length, 0, -1))  # Sorted in descending order
        arr_best_case = list(range(length))           # Already sorted
        arr_average_case = generate_random_array(length)  # Randomly generated array

        # Time analysis for worst case
        start_time = time.time()
        quicksort(arr_worst_case)
        end_time = time.time()
        time_worst_case.append(end_time - start_time)

        # Time analysis for best case
        start_time = time.time()
        quicksort(arr_best_case)
        end_time = time.time()
        time_best_case.append(end_time - start_time)

        # Time analysis for average case
        start_time = time.time()
        quicksort(arr_average_case)
        end_time = time.time()
        time_average_case.append(end_time - start_time)

    return time_worst_case, time_best_case, time_average_case

def plot_time_analysis(data_lengths, time_worst_case, time_best_case, time_average_case):
    plt.plot(data_lengths, time_worst_case, label='Worst Case')
    plt.plot(data_lengths, time_best_case, label='Best Case')
    plt.plot(data_lengths, time_average_case, label='Average Case')
    plt.xlabel('Data Length')
    plt.ylabel('Time (seconds)')
    plt.title('Quicksort Time Analysis')
    plt.legend()
    plt.grid(True)
    plt.show()

# Data lengths to analyze
data_lengths = [10, 50, 100, 500, 1000]

# Perform time analysis
time_worst_case, time_best_case, time_average_case = time_analysis(data_lengths)

# Plotting the results
plot_time_analysis(data_lengths, time_worst_case, time_best_case, time_average_case)
```
