```python
class UserInterface:
    def __init__(self):
        self.logged_in = False
        self.username = None

    def login(self, username, password):
        # Simulate login functionality
        if username == 'user' and password == 'pass':
            self.logged_in = True
            self.username = username
            return True
        else:
            return False

    def logout(self):
        # Simulate logout functionality
        self.logged_in = False
        self.username = None

    def check_system_status(self):
        # Usability Heuristic: Visibility of System Status
        if self.logged_in:
            print(f"User {self.username} is logged in.")
        else:
            print("User is not logged in.")

    def show_main_menu(self):
        # Usability Heuristic: Match between System and the Real World
        print("\nMain Menu:")
        print("1. View Profile")
        print("2. Edit Settings")
        print("3. Logout")

    def handle_user_input(self, choice):
        # Usability Heuristic: Recognition rather than Recall
        if choice == '1':
            self.view_profile()
        elif choice == '2':
            self.edit_settings()
        elif choice == '3':
            self.logout()

    def view_profile(self):
        # Usability Heuristic: Consistency and Standards
        if self.logged_in:
            print(f"Showing profile of {self.username}.")
        else:
            print("Please log in first.")

    def edit_settings(self):
        # Usability Heuristic: Flexibility and Efficiency of Use
        if self.logged_in:
            print(f"Editing settings for {self.username}.")
        else:
            print("Please log in first.")

    def handle_error(self):
        # Usability Heuristic: Help Users Recognize, Diagnose, and Recover from Errors
        print("An error occurred. Please try again or contact support.")

    def display_help(self):
        # Usability Heuristic: Help and Documentation
        print("\nHelp Documentation:")
        print("To view profile, select option 1.")
        print("To edit settings, select option 2.")
        print("To logout, select option 3.")

# Usage Example
ui = UserInterface()

# Simulate user interaction
username_input = input("Enter username: ")
password_input = input("Enter password: ")

if ui.login(username_input, password_input):
    ui.check_system_status()
    ui.show_main_menu()

    user_choice = input("Enter your choice: ")
    ui.handle_user_input(user_choice)
else:
    ui.handle_error()
    ui.display_help()
```

```python
import time

class UsabilityTester:
    def __init__(self, system):
        self.system = system
        self.users = []

    def add_user(self, user):
        self.users.append(user)

    def run_task_based_testing(self, task):
        print("Starting Task-Based Usability Testing...")
        for user in self.users:
            print(f"User {user.name} is performing task: {task}")
            start_time = time.time()
            # Simulate user performing the task
            self.system.execute_task(task)
            end_time = time.time()
            completion_time = end_time - start_time
            print(f"User {user.name} completed the task in {completion_time:.2f} seconds")
        print("Task-Based Usability Testing completed.")

    def run_think_aloud_protocol(self):
        print("Encouraging users to verbalize thoughts and actions during usability testing...")
        for user in self.users:
            print(f"User {user.name}: Please verbalize your thoughts while interacting with the system.")
            # Simulate user using the system and verbalizing actions
            self.system.simulate_user_interaction(user)
        print("Think-Aloud Protocol completed.")

    def conduct_surveys_and_questionnaires(self):
        print("Conducting Surveys and Questionnaires...")
        feedback = {}
        for user in self.users:
            print(f"User {user.name}: Please provide your feedback through the survey.")
            # Simulate collecting feedback from user
            feedback[user.name] = self.system.collect_feedback(user)
        print("Surveys and Questionnaires completed.")
        return feedback

# Example classes representing a system and user
class System:
    def execute_task(self, task):
        # Simulate executing a task
        time.sleep(2)  # Simulate task execution time

    def simulate_user_interaction(self, user):
        # Simulate user interaction and verbalization
        print(f"User {user.name}: I'm navigating through the system.")

    def collect_feedback(self, user):
        # Simulate collecting user feedback
        return input(f"User {user.name}: Please provide your feedback (e.g., rate usability): ")

class User:
    def __init__(self, name):
        self.name = name

# Usage Example
system = System()
tester = UsabilityTester(system)

# Add users for usability testing
user1 = User("Alice")
user2 = User("Bob")
tester.add_user(user1)
tester.add_user(user2)

# Perform usability testing using different principles
tester.run_task_based_testing("Login")
tester.run_think_aloud_protocol()
feedback_results = tester.conduct_surveys_and_questionnaires()

print("Usability Testing results:")
print(feedback_results)
```

```python
class UserCenteredDesign:
    def __init__(self):
        self.users = []

    def gather_user_requirements(self):
        print("Initiating user-centered approach: Gathering user requirements...")
        # Simulate gathering user requirements through interviews or surveys
        user_requirements = input("Please provide your requirements for the system: ")
        # Store user requirements for design phase
        self.users.append({'requirements': user_requirements})
        print("User requirements gathered successfully.")

    def design_interface(self):
        print("Designing user interface based on gathered requirements...")
        # Simulate designing user interface based on user requirements
        interface_design = "Mockup designs created based on user feedback"
        print("Interface design completed:", interface_design)

# Usage Example
ucd = UserCenteredDesign()

# Gather user requirements
ucd.gather_user_requirements()

# Design user interface based on requirements
ucd.design_interface()
```

```python
from collections import defaultdict

# Example classes representing a system and user
class System:
    def collect_feedback(self, user_requirements):
        # Simulate collecting user feedback based on requirements
        return input("Please provide your feedback (e.g., rate usability): ")

    def iterate_design(self, user_feedback):
        # Simulate updating design based on user feedback
        print("Updating design based on user feedback...")

class User:
    def __init__(self, name, requirements):
        self.name = name
        self.requirements = requirements

class UserCenteredDesign:
    def __init__(self):
        self.users = []
        self.system = System()

    def add_user(self, user):
        self.users.append(user)

    def gather_user_requirements(self):
        print("Initiating user-centered approach: Gathering user requirements...")
        for user in self.users:
            print(f"User {user.name}: {user.requirements}")
        print("User requirements gathered successfully.")

    def design_interface(self):
        print("Designing user interface based on gathered requirements...")
        # Simulate designing user interface based on user requirements
        interface_design = "Mockup designs created based on user feedback"
        print("Interface design completed:", interface_design)

class UserCenteredDesignWithTesting(UserCenteredDesign):
    def __init__(self):
        super().__init__()
        self.system = System()

    def conduct_empirical_testing(self):
        print("Initiating empirical testing...")
        user_feedback = defaultdict(list)
        for user in self.users:
            print(f"User {user.name} is testing the system.")
            feedback = self.system.collect_feedback(user.requirements)
            user_feedback[user.name].append(feedback)
        print("Empirical testing completed.")
        return user_feedback

# Usage Example
ucd_with_testing = UserCenteredDesignWithTesting()

# Add users with requirements for usability testing
user1 = User("Alice", "Login functionality")
user2 = User("Bob", "Profile management")
ucd_with_testing.add_user(user1)
ucd_with_testing.add_user(user2)

# Gather user requirements
ucd_with_testing.gather_user_requirements()

# Design user interface based on requirements
ucd_with_testing.design_interface()

# Conduct empirical testing with real users
ucd_with_testing.conduct_empirical_testing()
```

```python
class IterativeDesign:
    def __init__(self, system):
        self.system = system

    def iterate_design(self, user_feedback):
        print("Initiating iterative design...")
        # Simulate incorporating user feedback into design iterations
        for feedback in user_feedback:
            self.system.update_design(feedback)
        print("Iterative design completed.")

# Usage Example (continuation from UserCenteredDesignWithTesting)
class UserCenteredDevelopmentProcess(UserCenteredDesignWithTesting, IterativeDesign):
    def __init__(self):
        super().__init__()

    def execute_user_centered_development(self):
        # Execute user-centered development process
        self.gather_user_requirements()
        self.design_interface()
        self.conduct_empirical_testing()
        self.iterate_design()

# Usage Example
ucdp = UserCenteredDevelopmentProcess()
ucdp.execute_user_centered_development()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, utils, datasets

def build_model():
    """Build and compile a convolutional neural network model."""
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

def preprocess_data(train_images, train_labels, test_images, test_labels):
    """Preprocess MNIST data for training and testing."""
    # Reshape and scale pixel values to range [0, 1]
    train_images = train_images.reshape((-1, 28, 28, 1)).astype('float32') / 255.0
    test_images = test_images.reshape((-1, 28, 28, 1)).astype('float32') / 255.0
    return train_images, train_labels, test_images, test_labels

def train_model(model, train_images, train_labels, epochs=5, batch_size=32):
    """Train the convolutional neural network model."""
    history = model.fit(train_images, train_labels,
                        epochs=epochs,
                        batch_size=batch_size,
                        validation_split=0.1)
    return history

def evaluate_model(model, test_images, test_labels):
    """Evaluate the trained model on test data."""
    test_loss, test_accuracy = model.evaluate(test_images, test_labels)
    print(f'Test Accuracy: {test_accuracy:.4f}')
    return test_accuracy

def main():
    """Main function to build, train, and evaluate the CNN model."""
    # Load MNIST dataset
    (train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()

    # Preprocess data
    train_images, train_labels, test_images, test_labels = preprocess_data(train_images, train_labels, test_images, test_labels)

    # Build and compile the model
    model = build_model()

    # Train the model
    history = train_model(model, train_images, train_labels)

    # Evaluate the model on test data
    evaluate_model(model, test_images, test_labels)

if __name__ == '__main__':
    main()
```

#Different Measures for Evaluation (e.g., usefulness, efficiency, ease of learning, user satisfaction)
User Satisfaction Measurement in a Survey

```python
def conduct_user_survey():
    # Simulated user satisfaction survey responses
    responses = [
        {'satisfaction_score': 4},
        {'satisfaction_score': 3},
        {'satisfaction_score': 5},
        {'satisfaction_score': 2},
        {'satisfaction_score': 4}
    ]
    return responses

def measure_user_satisfaction(responses):
    if not responses:
        return 0  # No responses
    satisfaction_scores = [response['satisfaction_score'] for response in responses]
    average_satisfaction = sum(satisfaction_scores) / len(satisfaction_scores)
    return average_satisfaction

# Conduct the user survey and measure satisfaction
survey_responses = conduct_user_survey()
average_satisfaction_score = measure_user_satisfaction(survey_responses)
print(f"Average User Satisfaction Score: {average_satisfaction_score:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_survey_responses(num_responses, metric_name):
    """Simulate survey responses for a given evaluation metric."""
    if metric_name == 'usefulness':
        # Simulate usefulness scores (1 to 5)
        return np.random.randint(1, 6, size=num_responses)
    elif metric_name == 'efficiency':
        # Simulate efficiency scores (1 to 5)
        return np.random.randint(1, 6, size=num_responses)
    elif metric_name == 'ease_of_learning':
        # Simulate ease of learning scores (1 to 5)
        return np.random.randint(1, 6, size=num_responses)
    elif metric_name == 'user_satisfaction':
        # Simulate user satisfaction scores (1 to 5)
        return np.random.randint(1, 6, size=num_responses)
    else:
        raise ValueError("Invalid metric name. Supported metrics: usefulness, efficiency, ease_of_learning, user_satisfaction")

def plot_metric_histogram(metric_name, metric_scores):
    """Plot a histogram for a specific evaluation metric."""
    plt.figure(figsize=(8, 6))
    plt.hist(metric_scores, bins=np.arange(1, 7) - 0.5, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title(f'{metric_name.replace("_", " ").title()} Survey Results')
    plt.xlabel('Score')
    plt.ylabel('Number of Responses')
    plt.xticks(np.arange(1, 6))
    plt.yticks(np.arange(0, len(metric_scores) + 1, step=5))
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

def main():
    """Main function to simulate and plot multiple evaluation metrics."""
    num_responses = 100

    # Simulate survey responses for each metric
    usefulness_scores = simulate_survey_responses(num_responses, 'usefulness')
    efficiency_scores = simulate_survey_responses(num_responses, 'efficiency')
    ease_of_learning_scores = simulate_survey_responses(num_responses, 'ease_of_learning')
    user_satisfaction_scores = simulate_survey_responses(num_responses, 'user_satisfaction')

    # Plot histograms for each evaluation metric
    plot_metric_histogram('usefulness', usefulness_scores)
    plot_metric_histogram('efficiency', efficiency_scores)
    plot_metric_histogram('ease_of_learning', ease_of_learning_scores)
    plot_metric_histogram('user_satisfaction', user_satisfaction_scores)

if __name__ == '__main__':
    main()
```
