#Quantitative Evaluation Techniques (e.g., Keystroke-Level Evaluation)

Quantitative evaluation methods, like keystroke-level evaluation, provide objective measures of system performance and efficiency.



```python
class KeystrokeLevelModel:
    def __init__(self):
        # Define operation times in seconds (based on typical values)
        self.kb_stroke_time = 0.08  # Time to press or release a key (in seconds)
        self.mouse_click_time = 0.2  # Time to click a mouse button (in seconds)
        self.pointing_time = 1.1     # Time to move the mouse pointer to a target (in seconds)
        self.homogeneous_time = 0.4   # Time to move the hand between keyboard and mouse (in seconds)

    def calculate_klm_time(self, task_sequence):
        total_time = 0

        for operation in task_sequence:
            operation_type, operand = operation[0], operation[1]

            if operation_type == 'k':
                total_time += self.kb_stroke_time * operand  # Keystroke
            elif operation_type == 'p':
                total_time += self.pointing_time + self.mouse_click_time  # Pointing and clicking
            elif operation_type == 'h':
                total_time += self.homogeneous_time  # Hand movement between keyboard and mouse

        return total_time

# Example usage of the Keystroke-Level Model (KLM)
klm = KeystrokeLevelModel()

# Define a sequence of tasks as operations ('k' for keystrokes, 'p' for pointing, 'h' for hand movement)
task_sequence = [
    ('k', 5),  # Typing 5 characters
    ('h', 1),  # Moving hand from keyboard to mouse
    ('p', 1),  # Pointing and clicking once
    ('k', 8)   # Typing 8 more characters
]

# Calculate the estimated time using the KLM for the defined task sequence
estimated_time = klm.calculate_klm_time(task_sequence)

print(f"Estimated time to complete the task: {estimated_time:.2f} seconds")
```

#Userless Evaluation Using Qualitative and Quantitative Techniques (e.g., Expert Reviews, Heuristic Evaluation)

Evaluation without users involves techniques like expert reviews, heuristic evaluation, and guideline-based analysis to identify usability issues.

```python
import matplotlib.pyplot as plt
import numpy as np

class UserlessEvaluation:
    def __init__(self):
        # Define systems and their heuristic evaluation ratings
        self.systems_data = {
            'System A': {
                'Visibility of System Status': 2,
                'Match Between System and the Real World': 3,
                'User Control and Freedom': 4,
                'Consistency and Standards': 3,
                'Error Prevention': 2,
                'Recognition Rather than Recall': 4,
                'Flexibility and Efficiency of Use': 3,
                'Aesthetic and Minimalist Design': 4,
                'Help Users Recognize, Diagnose, and Recover from Errors': 2,
                'Help and Documentation': 1
            },
            'System B': {
                'Visibility of System Status': 3,
                'Match Between System and the Real World': 2,
                'User Control and Freedom': 3,
                'Consistency and Standards': 4,
                'Error Prevention': 3,
                'Recognition Rather than Recall': 3,
                'Flexibility and Efficiency of Use': 2,
                'Aesthetic and Minimalist Design': 3,
                'Help Users Recognize, Diagnose, and Recover from Errors': 2,
                'Help and Documentation': 3
            }
        }

        # Define weights for heuristic criteria (normalized to sum up to 1)
        self.heuristic_weights = {
            'Visibility of System Status': 0.1,
            'Match Between System and the Real World': 0.1,
            'User Control and Freedom': 0.1,
            'Consistency and Standards': 0.1,
            'Error Prevention': 0.05,
            'Recognition Rather than Recall': 0.1,
            'Flexibility and Efficiency of Use': 0.1,
            'Aesthetic and Minimalist Design': 0.15,
            'Help Users Recognize, Diagnose, and Recover from Errors': 0.05,
            'Help and Documentation': 0.1
        }

    def plot_evaluation_results(self):
        criteria = list(self.heuristic_weights.keys())
        weights = np.array(list(self.heuristic_weights.values()))

        for system_name, system_data in self.systems_data.items():
            # Prepare data for plotting
            ratings = np.array(list(system_data.values()))
            weighted_scores = np.dot(ratings, weights)

            # Plotting the evaluation results for the current system
            plt.figure(figsize=(10, 6))
            plt.bar(criteria, ratings, alpha=0.5, label='Ratings')
            plt.plot(criteria, weights, 'r-', linewidth=2, label='Weights')

            plt.xlabel('Heuristic Criteria')
            plt.ylabel('Rating / Weight')
            plt.title(f'Userless Evaluation Results for {system_name}')
            plt.xticks(rotation=45)
            plt.legend()

            plt.tight_layout()
            plt.show()

# Example usage of userless evaluation techniques with plotting
userless_evaluation = UserlessEvaluation()

# Plot the heuristic evaluation results for each system
userless_evaluation.plot_evaluation_results()
```
