#Use case diagram

```python
from graphviz import Digraph
from IPython.display import Image, display

# Create a new Digraph
dot = Digraph()

# Set global attributes for the diagram
dot.attr(rankdir='LR')  # Left to Right layout

# Add actors
dot.node('Customer', shape='actor', xlabel='Customer')
dot.node('Guest', shape='actor', xlabel='Guest')
dot.node('System', shape='ellipse', xlabel='Online Shopping System')

# Add use cases
use_cases = ['View Products', 'Add to Cart', 'Checkout', 'Register']
for use_case in use_cases:
    dot.node(use_case, shape='ellipse', style='filled', fillcolor='lightblue')

# Define associations
associations = {
    'Customer': ['View Products', 'Add to Cart', 'Checkout', 'Register'],
    'Guest': ['View Products', 'Checkout']
}

# Add edges for associations
for actor, use_case_list in associations.items():
    for use_case in use_case_list:
        dot.edge(actor, use_case)

# Add additional relationships with UML annotations
dot.edge('Customer', 'System', label='uses', style='dashed')
dot.edge('Guest', 'System', label='uses', style='dashed')

# Example of Aggregation (User + Profile)
dot.edge('Customer', 'Customer Profile', label='aggregates', arrowhead='odiamond')
dot.node('Customer Profile', shape='ellipse', style='filled', fillcolor='lightgreen')

# Example of Inheritance (Registered Customer inherits from Customer)
dot.edge('Customer', 'Registered Customer', label='inherits', arrowhead='onormal')
dot.node('Registered Customer', shape='actor', xlabel='Registered Customer')

# Render the diagram
diagram_file = 'use_case_diagram'
dot.render(diagram_file, format='png', cleanup=True)

# Display the diagram
display(Image(filename=f"{diagram_file}.png"))
```

#Context Diagram (System Context)

```python
from graphviz import Digraph
from IPython.display import Image, display

# Create a new Digraph for Context diagram
context = Digraph()

# Define system and external actors
context.node('User', shape='actor', xlabel='User')
context.node('System', shape='ellipse', xlabel='Online Shopping System')

# Define relationships (external interactions)
context.edge('User', 'System', label='Uses')

# Set layout and render the diagram
context.attr(rankdir='LR')  # Left to Right layout
context.render('context_diagram', format='png', cleanup=True)

# Display the diagram
display(Image(filename='context_diagram.png'))
```

#Container Diagram (High-Level Components)

```python
from graphviz import Digraph
from IPython.display import Image, display

# Create a new Digraph for Container diagram
container = Digraph()

# Define containers (Web Server, Application Server, Database)
container.node('Web Server', shape='component', xlabel='Web Server (Nginx)')
container.node('App Server', shape='component', xlabel='Application Server (Django)')
container.node('Database', shape='component', xlabel='Database (PostgreSQL)')

# Define relationships (interactions between containers)
container.edge('Web Server', 'App Server', label='Forwards HTTP requests')
container.edge('App Server', 'Database', label='Reads/Writes data')

# Set layout and render the diagram
container.attr(rankdir='LR')  # Left to Right layout
container.render('container_diagram', format='png', cleanup=True)

# Display the diagram
display(Image(filename='container_diagram.png'))
```

#Component Diagram (Detailed Components)

```python
# Create a new Digraph for Component diagram
component = Digraph()

# Define components (Authentication, Order Management, Product Catalog)
component.node('Auth Module', shape='component', xlabel='Authentication Module')
component.node('Order Service', shape='component', xlabel='Order Service')
component.node('Catalog Service', shape='component', xlabel='Product Catalog Service')

# Define relationships (interactions between components and database)
component.edge('Auth Module', 'Database', label='Accesses user data')
component.edge('Order Service', 'Database', label='Stores order data')
component.edge('Catalog Service', 'Database', label='Retrieves product data')

# Set layout and render the diagram
component.attr(rankdir='LR')  # Left to Right layout
component.render('component_diagram', format='png', cleanup=True)

# Display the diagram
display(Image(filename='component_diagram.png'))
```

#Code Diagram (Source Code Dependencies)

```python
from graphviz import Digraph
from IPython.display import Image, display

# Create a new Digraph for Code diagram
code = Digraph()

# Define code elements (User Model, Order Controller, Product Service)
code.node('User Model', shape='box', xlabel='User Model (user.py)')
code.node('Order Controller', shape='box', xlabel='Order Controller (order.py)')
code.node('Product Service', shape='box', xlabel='Product Service (product.py)')

# Define relationships (dependencies between code elements)
code.edge('Order Controller', 'User Model', label='Uses')
code.edge('Product Service', 'User Model', label='Uses')

# Set layout and render the diagram
code.attr(rankdir='LR')  # Left to Right layout
code.render('code_diagram', format='png', cleanup=True)

# Display the diagram
display(Image(filename='code_diagram.png'))
```
