```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate random data for demonstration
np.random.seed(0)
months = np.arange(1, 13)
revenue = np.random.randint(20000, 80000, size=12)
expenses = np.random.randint(15000, 60000, size=12)
profit = revenue - expenses

# Set style
plt.style.use('seaborn-dark')

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

# Plot 1 - Revenue trend
axes[0, 0].plot(months, revenue, color='blue', marker='o', linestyle='-')
axes[0, 0].set_title('Revenue Trend', fontsize=16)
axes[0, 0].set_xlabel('Month', fontsize=12)
axes[0, 0].set_ylabel('Revenue ($)', fontsize=12)

# Plot 2 - Expenses trend
axes[0, 1].plot(months, expenses, color='red', marker='s', linestyle='--')
axes[0, 1].set_title('Expenses Trend', fontsize=16)
axes[0, 1].set_xlabel('Month', fontsize=12)
axes[0, 1].set_ylabel('Expenses ($)', fontsize=12)

# Plot 3 - Profit trend
axes[1, 0].plot(months, profit, color='green', marker='^', linestyle='-')
axes[1, 0].set_title('Profit Trend', fontsize=16)
axes[1, 0].set_xlabel('Month', fontsize=12)
axes[1, 0].set_ylabel('Profit ($)', fontsize=12)

# Plot 4 - Summary
summary_labels = ['Revenue', 'Expenses', 'Profit']
summary_data = [revenue.sum(), expenses.sum(), profit.sum()]
axes[1, 1].bar(summary_labels, summary_data, color=['blue', 'red', 'green'])
axes[1, 1].set_title('Summary', fontsize=16)
axes[1, 1].set_ylabel('Amount ($)', fontsize=12)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
