```python
class Graph:
    def __init__(self):
        self.adjacency_list = {}

    def add_vertex(self, vertex):
        if vertex not in self.adjacency_list:
            self.adjacency_list[vertex] = []

    def add_edge(self, vertex1, vertex2):
        # Adding edge between vertex1 and vertex2 (undirected)
        if vertex1 in self.adjacency_list and vertex2 in self.adjacency_list:
            self.adjacency_list[vertex1].append(vertex2)
            self.adjacency_list[vertex2].append(vertex1)

    def show_connections(self):
        for vertex, connections in self.adjacency_list.items():
            print(f"{vertex} --> {connections}")

# Example usage:
if __name__ == "__main__":
    # Create a graph instance
    graph = Graph()

    # Add vertices
    graph.add_vertex('A')
    graph.add_vertex('B')
    graph.add_vertex('C')
    graph.add_vertex('D')

    # Add edges
    graph.add_edge('A', 'B')
    graph.add_edge('A', 'C')
    graph.add_edge('B', 'D')
    graph.add_edge('C', 'D')

    # Display connections
    graph.show_connections()
```

#Graph Search Algorithms (BFS and DFS)

Breadth-First Search (BFS):


```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

def bfs(graph, start_node):
    visited = set()
    queue = deque([start_node])
    visited.add(start_node)

    while queue:
        node = queue.popleft()
        print(node, end=" ")  # Process node (e.g., print or store in a list)

        for neighbor in graph.neighbors(node):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

# Create a sample graph (undirected)
G = nx.Graph()
edges = [(1, 2), (1, 3), (2, 4), (2, 5), (3, 6), (4, 7), (5, 8)]
G.add_edges_from(edges)

# Perform BFS starting from node 1
print("BFS traversal:")
bfs(G, 1)
```

Depth-First Search (DFS):


```python
def dfs(graph, start_node, visited=None):
    if visited is None:
        visited = set()
    visited.add(start_node)
    print(start_node, end=" ")  # Process node (e.g., print or store in a list)

    for neighbor in graph.neighbors(start_node):
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

# Create a sample graph (undirected)
G = nx.Graph()
edges = [(1, 2), (1, 3), (2, 4), (2, 5), (3, 6), (4, 7), (5, 8)]
G.add_edges_from(edges)

# Perform DFS starting from node 1
print("DFS traversal:")
dfs(G, 1)
```

#DFS trees

```python
import networkx as nx
import matplotlib.pyplot as plt

def dfs_tree(graph, start_node):
    # Initialize empty DFS tree
    dfs_tree = nx.DiGraph()

    # Helper function to perform DFS and construct tree
    def dfs_visit(node, parent=None):
        if node not in dfs_tree:
            dfs_tree.add_node(node)
            if parent is not None:
                dfs_tree.add_edge(parent, node)
            for neighbor in graph.neighbors(node):
                if neighbor not in dfs_tree:
                    dfs_visit(neighbor, node)

    # Start DFS traversal from the specified node
    dfs_visit(start_node)

    return dfs_tree

# Create a sample graph (undirected)
G = nx.Graph()
edges = [(1, 2), (1, 3), (2, 4), (2, 5), (3, 6), (4, 7), (5, 8)]
G.add_edges_from(edges)

# Specify the root node for DFS
root_node = 1

# Construct DFS tree starting from the root node
dfs_tree = dfs_tree(G, root_node)

# Draw the original graph and the DFS tree for visualization
plt.figure(figsize=(12, 6))

plt.subplot(121)
nx.draw(G, with_labels=True, node_color='skyblue', node_size=800, font_size=16, font_color='black', font_weight='bold')
plt.title("Original Graph")

plt.subplot(122)
nx.draw(dfs_tree, with_labels=True, node_color='lightgreen', node_size=800, font_size=16, font_color='black', font_weight='bold', arrows=True)
plt.title("DFS Tree (Rooted at {})".format(root_node))

plt.tight_layout()
plt.show()
```

```python
import networkx as nx

def bellman_ford(graph, source):
    # Initialize distances from source to all other nodes
    distances = {node: float('inf') for node in graph.nodes()}
    distances[source] = 0

    # Relax edges repeatedly
    for _ in range(len(graph.nodes()) - 1):
        for u, v, data in graph.edges(data=True):
            if distances[u] + data['weight'] < distances[v]:
                distances[v] = distances[u] + data['weight']

    # Check for negative cycles
    for u, v, data in graph.edges(data=True):
        if distances[u] + data['weight'] < distances[v]:
            raise ValueError("Graph contains a negative-weight cycle")

    return distances

# Example usage:
G = nx.DiGraph()
G.add_weighted_edges_from([(1, 2, 1), (1, 3, 4), (2, 3, 2), (2, 4, 5), (3, 4, 1)])

source_node = 1
shortest_distances = bellman_ford(G, source_node)
print("Shortest distances from node", source_node, ":")
print(shortest_distances)
```

```python
import heapq

def dijkstra(graph, source):
    distances = {node: float('inf') for node in graph.nodes()}
    distances[source] = 0
    priority_queue = [(0, source)]  # (distance, node)

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight['weight']

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances

# Example usage:
G = nx.Graph()
G.add_weighted_edges_from([(1, 2, 1), (1, 3, 4), (2, 3, 2), (2, 4, 5), (3, 4, 1)])

source_node = 1
shortest_distances = dijkstra(G, source_node)
print("Shortest distances from node", source_node, ":")
print(shortest_distances)
```

```python
def floyd_warshall(graph):
    nodes = list(graph.nodes())
    num_nodes = len(nodes)

    distances = {node1: {node2: float('inf') for node2 in nodes} for node1 in nodes}

    for node in nodes:
        distances[node][node] = 0

    for u, v, data in graph.edges(data=True):
        distances[u][v] = min(distances[u][v], data['weight'])

    for k in nodes:
        for i in nodes:
            for j in nodes:
                if distances[i][j] > distances[i][k] + distances[k][j]:
                    distances[i][j] = distances[i][k] + distances[k][j]

    return distances

# Example usage:
G = nx.Graph()
G.add_weighted_edges_from([(1, 2, 1), (1, 3, 4), (2, 3, 2), (2, 4, 5), (3, 4, 1)])

all_pairs_shortest_distances = floyd_warshall(G)
print("All pairs shortest distances:")
print(all_pairs_shortest_distances)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def topological_sort(graph):
    # Calculate in-degree for each node
    in_degree = {node: 0 for node in graph.nodes()}
    for u, v in graph.edges():
        in_degree[v] += 1

    # Initialize a queue with nodes having zero in-degree
    zero_in_degree_queue = [node for node, degree in in_degree.items() if degree == 0]
    topological_order = []

    while zero_in_degree_queue:
        node = zero_in_degree_queue.pop(0)
        topological_order.append(node)

        for neighbor in graph.neighbors(node):
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                zero_in_degree_queue.append(neighbor)

    # Check if all nodes have been visited
    if len(topological_order) != len(graph.nodes()):
        raise ValueError("Graph contains a cycle (not a DAG)")

    return topological_order

# Example usage:
G = nx.DiGraph()
edges = [(1, 2), (1, 3), (2, 4), (3, 4), (3, 5), (4, 6), (5, 6)]
G.add_edges_from(edges)

try:
    sorted_order = topological_sort(G)
    print("Topological order of nodes:", sorted_order)
except ValueError as e:
    print(e)

# Draw the DAG for visualization
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=800, font_size=16, font_color='black', font_weight='bold', arrows=True)
plt.title("Directed Acyclic Graph (DAG)")
plt.show()
```

#Dynamic Programming on Directed Acyclic Graphs (DAGs)
Dynamic programming (DP) on DAGs involves solving optimization problems by breaking them down into smaller subproblems and utilizing the acyclic nature of the graph to efficiently compute results.



```python
def dp_on_dag(graph, source):
    # Perform topological sorting
    sorted_order = topological_sort(graph)

    # Initialize DP table
    dp = {node: float('-inf') for node in graph.nodes()}
    dp[source] = 0

    # Compute DP values using topological order
    for node in sorted_order:
        if dp[node] != float('-inf'):  # Process only reachable nodes
            for neighbor in graph.neighbors(node):
                dp[neighbor] = max(dp[neighbor], dp[node] + graph[node][neighbor]['weight'])

    return dp

# Example usage:
G = nx.DiGraph()
edges = [(1, 2, {'weight': 5}), (1, 3, {'weight': 3}), (2, 4, {'weight': 7}), (3, 4, {'weight': 4}), (3, 5, {'weight': 2}), (4, 6, {'weight': 6}), (5, 6, {'weight': 8})]
G.add_edges_from(edges)

source_node = 1
dp_values = dp_on_dag(G, source_node)
print("Maximum path values from source node", source_node, ":")
print(dp_values)
```

#Minimum Spanning Trees (MSTs)

image.png

```python
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n

    def find(self, u):
        if self.parent[u] != u:
            self.parent[u] = self.find(self.parent[u])  # path compression
        return self.parent[u]

    def union(self, u, v):
        root_u = self.find(u)
        root_v = self.find(v)

        if root_u != root_v:
            # Union by rank
            if self.rank[root_u] > self.rank[root_v]:
                self.parent[root_v] = root_u
            elif self.rank[root_u] < self.rank[root_v]:
                self.parent[root_u] = root_v
            else:
                self.parent[root_v] = root_u
                self.rank[root_u] += 1

def kruskal_mst(graph):
    # Collect all edges and sort them by weight
    edges = []
    for u, v, data in graph.edges(data=True):
        edges.append((data['weight'], u, v))
    edges.sort()  # sort by weight

    # Initialize Union-Find structure
    uf = UnionFind(len(graph.nodes()))
    mst_edges = []
    mst_weight = 0

    # Process edges in increasing order of weight
    for weight, u, v in edges:
        if uf.find(u) != uf.find(v):
            uf.union(u, v)
            mst_edges.append((u, v))
            mst_weight += weight

    return mst_edges, mst_weight

# Example usage:
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
edges = [
    (0, 1, {'weight': 4}),
    (0, 7, {'weight': 8}),
    (1, 2, {'weight': 8}),
    (1, 7, {'weight': 11}),
    (2, 3, {'weight': 7}),
    (2, 8, {'weight': 2}),
    (2, 5, {'weight': 4}),
    (3, 4, {'weight': 9}),
    (3, 5, {'weight': 14}),
    (4, 5, {'weight': 10}),
    (5, 6, {'weight': 2}),
    (6, 7, {'weight': 1}),
    (6, 8, {'weight': 6}),
    (7, 8, {'weight': 7})
]
G.add_edges_from(edges)

mst_edges, mst_weight = kruskal_mst(G)
print("Minimum Spanning Tree (Kruskal's Algorithm):")
print("Edges:", mst_edges)
print("Total Weight:", mst_weight)

# Draw the original graph and MST for visualization
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=True, node_color='lightblue', node_size=800, font_size=16, font_color='black', font_weight='bold', arrows=False)
nx.draw_networkx_edges(G, pos, edgelist=mst_edges, edge_color='red', width=2)
plt.title("Minimum Spanning Tree (Kruskal's Algorithm)")
plt.show()
```

```python
import heapq

def prim_mst(graph):
    mst_edges = []
    mst_weight = 0
    visited = set()
    start_node = next(iter(graph.nodes()))  # start from any node

    priority_queue = [(0, start_node)]  # (weight, node)

    while priority_queue:
        weight, u = heapq.heappop(priority_queue)
        if u not in visited:
            visited.add(u)
            mst_weight += weight

            for v, data in graph[u].items():
                if v not in visited:
                    heapq.heappush(priority_queue, (data['weight'], v))
                    mst_edges.append((u, v))

    return mst_edges, mst_weight

# Example usage:
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
edges = [
    (0, 1, {'weight': 4}),
    (0, 7, {'weight': 8}),
    (1, 2, {'weight': 8}),
    (1, 7, {'weight': 11}),
    (2, 3, {'weight': 7}),
    (2, 8, {'weight': 2}),
    (2, 5, {'weight': 4}),
    (3, 4, {'weight': 9}),
    (3, 5, {'weight': 14}),
    (4, 5, {'weight': 10}),
    (5, 6, {'weight': 2}),
    (6, 7, {'weight': 1}),
    (6, 8, {'weight': 6}),
    (7, 8, {'weight': 7})
]
G.add_edges_from(edges)

mst_edges, mst_weight = prim_mst(G)
print("Minimum Spanning Tree (Prim's Algorithm):")
print("Edges:", mst_edges)
print("Total Weight:", mst_weight)

# Draw the original graph and MST for visualization
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=True, node_color='lightblue', node_size=800, font_size=16, font_color='black', font_weight='bold', arrows=False)
nx.draw_networkx_edges(G, pos, edgelist=mst_edges, edge_color='red', width=2)
plt.title("Minimum Spanning Tree (Prim's Algorithm)")
plt.show()
```

#Strongly connected directed graphs (SCCs)

image.png

```python
import networkx as nx

def kosaraju_scc(graph):
    # Step 1: Perform DFS on the original graph and record finishing times
    def dfs_visit(node, visited, stack):
        visited.add(node)
        for neighbor in graph.neighbors(node):
            if neighbor not in visited:
                dfs_visit(neighbor, visited, stack)
        stack.append(node)

    # Step 2: Transpose the graph
    transposed_graph = graph.reverse()

    # Step 3: Perform DFS on the transposed graph in the order of finishing times
    def dfs_scc(node, visited, component):
        visited.add(node)
        component.append(node)
        for neighbor in transposed_graph.neighbors(node):
            if neighbor not in visited:
                dfs_scc(neighbor, visited, component)

    # Step 4: Find SCCs using Kosaraju's algorithm
    visited = set()
    stack = []

    # Perform DFS on the original graph to get finishing times
    for node in graph.nodes():
        if node not in visited:
            dfs_visit(node, visited, stack)

    # Perform DFS on the transposed graph in the order of finishing times
    visited.clear()
    scc_list = []

    while stack:
        node = stack.pop()
        if node not in visited:
            component = []
            dfs_scc(node, visited, component)
            scc_list.append(component)

    return scc_list

# Example usage:
G = nx.DiGraph()
edges = [(1, 2), (2, 3), (3, 1), (3, 4), (4, 5), (5, 6), (6, 4), (7, 6), (7, 8), (8, 7)]
G.add_edges_from(edges)

scc_list = kosaraju_scc(G)
print("Strongly Connected Components (SCCs):")
for component in scc_list:
    print(component)
```

```python
from collections import defaultdict

def two_sat_scc(graph):
    # Step 1: Find SCCs using Kosaraju's algorithm
    scc_list = kosaraju_scc(graph)

    # Step 2: Check for satisfiability
    variable_assignment = {}
    for component in scc_list:
        for node in component:
            variable = abs(node)
            value = (node > 0)
            if variable not in variable_assignment:
                variable_assignment[variable] = value
            elif variable_assignment[variable] != value:
                return None  # Conflicting assignment detected

    # If no conflicts, return variable assignments
    return variable_assignment

# Example usage:
# Construct 2-SAT instance as a directed graph
G = nx.DiGraph()
# Example formula: (x1 or x2) and (not x1 or x3) and (not x2 or x4) and (not x3 or not x4)
G.add_edges_from([(1, 2), (-1, 3), (-2, 4), (-3, -4)])

variable_assignment = two_sat_scc(G)
if variable_assignment is not None:
    print("Satisfiable Assignment:")
    for variable, value in variable_assignment.items():
        print(f"x{variable} = {value}")
else:
    print("No satisfying assignment found.")
```

```python
from collections import defaultdict, deque

class Graph:
    def __init__(self, vertices):
        self.vertices = vertices
        self.adjacency_list = defaultdict(list)

    def add_edge(self, u, v, capacity):
        self.adjacency_list[u].append({'v': v, 'capacity': capacity, 'flow': 0})
        self.adjacency_list[v].append({'v': u, 'capacity': 0, 'flow': 0})  # Adding reverse edge

    def bfs(self, source, sink, parent):
        visited = [False] * self.vertices
        queue = deque([source])
        visited[source] = True

        while queue:
            u = queue.popleft()

            for edge in self.adjacency_list[u]:
                v = edge['v']
                if not visited[v] and edge['capacity'] > edge['flow']:  # Only consider if there's remaining capacity
                    queue.append(v)
                    visited[v] = True
                    parent[v] = u
                    if v == sink:
                        return True
        return False

    def ford_fulkerson(self, source, sink):
        parent = [-1] * self.vertices
        max_flow = 0

        while self.bfs(source, sink, parent):
            path_flow = float('Inf')
            s = sink

            while s != source:
                for edge in self.adjacency_list[parent[s]]:
                    if edge['v'] == s:
                        path_flow = min(path_flow, edge['capacity'] - edge['flow'])
                s = parent[s]

            # Update residual capacities of the edges and reverse edges along the path
            v = sink
            while v != source:
                u = parent[v]
                for edge in self.adjacency_list[u]:
                    if edge['v'] == v:
                        edge['flow'] += path_flow
                        for reverse_edge in self.adjacency_list[v]:
                            if reverse_edge['v'] == u:
                                reverse_edge['flow'] -= path_flow
                                break
                        break
                v = parent[v]

            max_flow += path_flow

        return max_flow

# Example usage:
g = Graph(6)  # Initialize graph with 6 vertices
g.add_edge(0, 1, 16)
g.add_edge(0, 2, 13)
g.add_edge(1, 2, 10)
g.add_edge(1, 3, 12)
g.add_edge(2, 1, 4)
g.add_edge(2, 4, 14)
g.add_edge(3, 2, 9)
g.add_edge(3, 5, 20)
g.add_edge(4, 3, 7)
g.add_edge(4, 5, 4)

source = 0
sink = 5
print("Maximum flow:", g.ford_fulkerson(source, sink))
```

```python
class BipartiteGraph:
    def __init__(self, left_set, right_set):
        self.left_set = left_set
        self.right_set = right_set
        self.adjacency_list = defaultdict(list)
        self.pair_u = {}
        self.pair_v = {}
        self.dist = {}

    def add_edge(self, u, v):
        self.adjacency_list[u].append(v)

    def bfs(self):
        queue = deque()
        for u in self.left_set:
            if self.pair_u[u] is None:
                self.dist[u] = 0
                queue.append(u)
            else:
                self.dist[u] = float('inf')

        self.dist[None] = float('inf')

        while queue:
            u = queue.popleft()
            if self.dist[u] < self.dist[None]:
                for v in self.adjacency_list[u]:
                    if self.dist[self.pair_v[v]] == float('inf'):
                        self.dist[self.pair_v[v]] = self.dist[u] + 1
                        queue.append(self.pair_v[v])

        return self.dist[None] != float('inf')

    def dfs(self, u):
        if u is not None:
            for v in self.adjacency_list[u]:
                if self.dist[self.pair_v[v]] == self.dist[u] + 1:
                    if self.dfs(self.pair_v[v]):
                        self.pair_v[v] = u
                        self.pair_u[u] = v
                        return True
            self.dist[u] = float('inf')
            return False
        return True

    def hopcroft_karp(self):
        self.pair_u = {u: None for u in self.left_set}
        self.pair_v = {v: None for v in self.right_set}
        matching = 0

        while self.bfs():
            for u in self.left_set:
                if self.pair_u[u] is None:
                    if self.dfs(u):
                        matching += 1

        return matching

# Example usage:
left_set = {1, 2, 3}
right_set = {4, 5, 6}
bipartite_graph = BipartiteGraph(left_set, right_set)
bipartite_graph.add_edge(1, 4)
bipartite_graph.add_edge(1, 5)
bipartite_graph.add_edge(2, 4)
bipartite_graph.add_edge(3, 6)

print("Maximum matching size:", bipartite_graph.hopcroft_karp())
```
